Агент 5 (играет то, что сыграл противник в прошлом ходу, если ход первый-то рэндом)

```python
%%writefile fifth_agent.py

import random

def fifth_agent(observation, configuration):
    if observation.step > 0:
        return observation.lastOpponentAction
    else:
        return random.randrange(0, configuration.signs)
```

Турнир между агентами

```python
agents = ['first_agent.py',
          'second_agent.py',
          'third_agent.py',
          'fourth_agent.py',
          'fifth_agent.py',
          'sixth_agent.py',
          'seventh_agent.py',
          'eightth_agent.py']


rank = {agent: 0 for agent in agents}

for times in range(5):
          
    for i in range(len(agents)):
        for j in range(i+1, len(agents)):
            e = evaluate(
            "rps", 
            [agents[i], agents[j]], 
            configuration={"episodeSteps": 100, 'tieRewardThreshold': 1} )
            if e[0][0] > e[0][1]:
                rank[agents[i]] += 1
            elif e[0][0] < e[0][1]:
                rank[agents[j]] += 1
rank_se = pd.Series(rank, name="Турнирная таблица").sort_values(ascending=False).to_frame()
rank_se
```
